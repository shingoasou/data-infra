```python
import os
import glob
import random
import shutil
from icrawler.builtin import GoogleImageCrawler
```

```python
root_dir = 'fish_images/'
```

```python
fish_names = ['アイゴ', 'オニカサゴ', 'カワハギ', 'キュウセンベラ', 'クサフグ', 'ソウシハギ', 'マハゼ', 'マアジ', 'マイワシ', 'ミノカサゴ', 'メジナ', 'メバル']
```

```python
data_count = 100
```

```python
def image_crawler(fish_name, data_count, root_dir):
    crawler = GoogleImageCrawler(storage={'root_dir':root_dir + fish_name + '/train'})
    filters = dict(size='large', type='photo')
    crawler.crawl(keyword=fish_names, filters=filters, max_num=data_count)
    
    if os.path.isdir(root_dir + fish_name + '/test'):
        shutil.rmtree(root_dir + fish_name + '/test')
    os.makedirs(root_dir + fish_name + '/test')
    
    file_list = glob.glob(root_dir + fish_name + '/train/*')
    test_ratio = 0.2
    test_files = random.sample(file_list, int(len(file_list) * test_ratio))
    
    for test_file in test_files:
        shutil.move(test_file, root_dir + fish_name + '/test/')
```

```python
for fish_name in fish_names:
    image_crawler(fish_name, data_count, root_dir)
```
